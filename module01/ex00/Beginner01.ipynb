```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# Load the dataset using numpy
data = np.genfromtxt('Salary dataset.csv', delimiter=',', skip_header=1)

# Separate the columns into YearsExperience and Salary
years_experience = data[:, 1]
salary = data[:, 2]
```

```python
# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(years_experience, salary)

# Add labels and title
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Scatter plot of YearsExperience vs. Salary')

# Show plot
plt.tight_layout()
plt.show()
```
