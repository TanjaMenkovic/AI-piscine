```python
df = pd.read_csv('Salary dataset.csv')
df.head()
```

Displaying data using scatterplot

```python
# Scatter plot with marginal boxplots
sns.set_theme(style="darkgrid")

# Create a JointGrid for adding boxplots on the margins
g = sns.JointGrid(data=df, x='YearsExperience', y='Salary', space=0)

# Add the scatterplot with costum purple colot
g.plot(sns.scatterplot, sns.histplot, color="#6A0DAD")

# Show plot
g.fig.tight_layout()
g.fig.subplots_adjust(top=0.95, right=0.95)
```
