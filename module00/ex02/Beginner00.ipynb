```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

# EX02

```python
df = pd.read_csv('Salary dataset.csv')
df.head()
```

Training my first model using linear regression

```python
# Check for NaN values in the entire DataFrame
nan_values = df.isna().any().any()

if nan_values:
    print("There are NaN values in the DataFrame")
else:
    print("There are no NaN values in the DataFrame")

# Eliminating NaN or missing input numbers
df.fillna(method ='ffill', inplace = True)

# Define features (X) and target (y)
X = df[['YearsExperience']]
y = df['Salary']

# Ensure X has valid feature names (column names)
X.columns = ['YearsExperience']  # Assigning column names if not already present

# Create and train the linear regression model with scaled features
model = LinearRegression()
model.fit(X, y)
```

```python
# Add predictions to the DataFrame
df['Predicted_Salary'] = model.predict(df[['YearsExperience']])
```

Predicting salaries based on years of experience

```python
# Function to predict salary based on years of experience input
def predict_salary(years_of_experience):
    predicted_salary = model.predict(years_of_experience)
    return predicted_salary[0]

while 1:
  # Get input from user in Google Colab
  years_of_experience_input = input("Enter years of experience (or anything that is not a number for exiting): ")

  if not years_of_experience_input.isdigit():
    break

  input_df = pd.DataFrame({'YearsExperience': [float(years_of_experience_input)]})

  # Predict salary based on user input
  predicted_salary = predict_salary(input_df)

  # Print the predicted salary with correct formatting
  print(f"Predicted salary for {years_of_experience_input} years of experience : {predicted_salary}")
```
