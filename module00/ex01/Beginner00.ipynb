```python
import pandas as pd
import seaborn as sns
from sklearn import preprocessing, svm
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

# EX01

```python
df = pd.read_csv('Salary dataset.csv')
df.head()
```

Training my first model using linear regression

```python
# Check for NaN values in the entire DataFrame
nan_values = df.isna().any().any()

if nan_values:
    print("There are NaN values in the DataFrame")
else:
    print("There are no NaN values in the DataFrame")

# Eliminating NaN or missing input numbers
df.fillna(method ='ffill', inplace = True)

# Define features (X) and target (y)
X = df[['YearsExperience']]
y = df['Salary']

# Ensure X has valid feature names (column names)
X.columns = ['YearsExperience']  # Assigning column names if not already present

# Create and train the linear regression model with scaled features
model = LinearRegression()
model.fit(X, y)
```

```python
# Add predictions to the DataFrame
df['Predicted_Salary'] = model.predict(df[['YearsExperience']])
df.tail()
```

Displaying linear regression bar on my first graph

```python
# Scatter plot with marginal boxplots
sns.set_theme(style="darkgrid")

# Scatter plot with regression line
plot = sns.lmplot(x='YearsExperience', y='Salary',
           data=df, line_kws={'color': 'blue'},
           scatter_kws={'color': "blue"}
           )
plot.fig.suptitle("Linear Regression: Years of Experience vs Salary", fontsize=10)
plot.set_axis_labels("Years of Experience", "Salary")

# Set axis sizes
plot.set_xlabels(fontsize=10)
plot.set_ylabels(fontsize=10)

# Adjust subplots so that titles don't overlap
plot.fig.subplots_adjust(top = 0.92)

# Adjust figure size to make x-axis longer than y-axis
plot.fig.set_size_inches(10, 6)

display(plot)
```
